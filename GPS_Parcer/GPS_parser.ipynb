```python
>>> from serial import Serial
>>> from pynmeagps import NMEAReader
>>> stream = Serial('/dev/cu.usbmodem1401', 9600, timeout=3)
while(1):
    nmr = NMEAReader(stream)
    (raw_data, parsed_data) = nmr.read()
    msg = NMEAReader.parse(raw_data)
    print(raw_data)
    print(parsed_data)
    print(msg.msgID)
    if msg.msgID == 'RMC':
        print(msg.status, msg.date, msg.time, msg.lat, msg.lon, msg.spd)
        
```
